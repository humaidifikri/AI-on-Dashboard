```python
import pandas as pd

from sklearn.preprocessing import StandardScaler,OneHotEncoder,MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split,cross_val_score

from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

import optuna
```

```python
df = pd.read_csv('../data/weight_change_dataset.csv')
df['Current Weight (lbs)'] = df['Current Weight (lbs)'].apply(lambda x: x/2.2).round(2)
df['Final Weight (lbs)'] = df['Final Weight (lbs)'].apply(lambda x: x/2.2).round(2)

X = df.drop(columns=['Participant ID','Weight Change (lbs)','Duration (weeks)','Final Weight (lbs)'])
y = df['Final Weight (lbs)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle=True, random_state=42)

num_cols = X.select_dtypes(include=['float64','int64'],exclude='object').columns
cat_cols = X.select_dtypes(exclude=['float64','int64'],include='object').columns

num_pipe = Pipeline([
    ('scl',StandardScaler()),
])

cat_pipe = Pipeline([
    ('enc',OneHotEncoder()),
])

transformer = ColumnTransformer([
    ('num',num_pipe,num_cols),
    ('cat',cat_pipe,cat_cols)
])

def objective(trial):
    # Hyperparameters yang dituning
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 350, 550),
        "max_depth": trial.suggest_int("max_depth", 6, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.25, 0.35),
        "subsample": trial.suggest_float("subsample", 0.8, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.85, 0.95),
        "gamma": trial.suggest_float("gamma", 0, 2),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 3)
    }
    
    # Model pipeline baru dengan parameter hasil trial
    model = Pipeline([
        ("trf", transformer), 
        ("classifier", XGBRegressor(
            **params
        ))
    ])
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    return mae
# Jalankan study Optuna
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=1000)

# Lihat hasil terbaik
print("Best params:", study.best_params)
print("Best score:", study.best_value)

optuna.visualization.matplotlib.plot_slice(study);
```
